```python
import pandas as pd
data = pd.read_csv('data_files/Evaluation_DF_v3.csv', encoding='utf-8')
data.columns.values[0] = 'Date'
data = data.set_index('Date')
print (data.head(5))
```

```python
data = data.drop(columns=['Position', 'Expectation','-Qt'])
print (data.head())
```

```python
print(data.describe())
```

```python
#target (dependent variable or y) is Direction
#features (independent variables or x) are ADF, ExClosemomenums, expectation, error, Qt
import numpy as np
labels = np.array(data['Direction'])

data = data.drop('Direction', axis=1)

features_list = list(data.columns)

features = np.array(data)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, 
                        labels, test_size = 192, shuffle=False, stratify=None)
```

```python
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 50, random_state=3)
rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
print (predictions)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
print(metrics.classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
#Confusion Matrix
pd.crosstab(y_test, predictions, rownames = ['True'], colnames=['Predicted'], margins=True)
```

```python
rndstates = [x for x in range(0,50)]
trees = list(range(10, 200, 10))
criterions = ['gini', 'entropy']
acc_score = []
#print (trees)
for criterion in criterions:
    for j in trees:
        for i in rndstates:
            rf = RandomForestClassifier(criterion=criterion, n_estimators = j, random_state=i)
            rf.fit(x_train, y_train)
            predictions = rf.predict(x_test)
            acc_score.append(accuracy_score(y_test,predictions))
            print(f'Criterion {criterion}, Tree {j} with Random state {i} has accuracy of {accuracy_score(y_test, predictions)}')
    
```

```python
max(acc_score)
```

```python
rf_2 = RandomForestClassifier(criterion ='entropy', n_estimators = 20, random_state=18)
rf_2.fit(x_train, y_train)
predictions = rf_2.predict(x_test)
print (predictions)
print(accuracy_score(y_test, predictions))
```
